# Pandas Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
df3.plot.scatter(x='a',y='b', color='red', s=50, figsize=(12,3))
```

```python
df3['a'].plot.hist(bins=10)
```

```python
plt.style.use('ggplot')
import matplotlib as mpl
```

```python
mpl.rcParams['patch.force_edgecolor'] = True
df3['a'].plot.hist(bins=25, alpha=.5)
```

```python
df3[['a','b']].plot.box() # Can also pass a by= argument for groupby
```

```python
df3['d'].plot.kde()
```

```python
df3['d'].plot.kde(lw=5, ls='--')
```
