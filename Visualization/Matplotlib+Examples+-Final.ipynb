# Matplotlib Exercises 

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'blue')
axes.set_xlabel('X') # Notice the use of set_ to begin methods
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,.2,.2]) # inset axes
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,.2,.2]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'red')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

# Insert Figure Axes 2
axes2.plot(x, y, 'red')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
#axes2.set_ylim([0, 200])
#axes2.set_xlim([2, 100])
```
