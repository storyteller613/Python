## Export Excel File

```python
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"

baby_name = pd.read_csv(URL)
baby_name.head()
```

```python
girls = baby_name[baby_name['Gender']=="FEMALE"]
boys = baby_name[baby_name['Gender']=="MALE"]
```

```python
excel_file = pd.ExcelWriter("Baby Name.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name="Girls", index=False)
boys.to_excel(excel_file, sheet_name="Boys", index=False, columns=["Gender", 'Child\'s First Name', 'Count'])
```

```python
excel_file.save()
```
