## Changing pandas Options with Attributes and Dot Syntax

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df
```

```python
pd.options.display.max_rows = 4 #max number of rows displayed
```

```python
df
```

```python
pd.options.display.max_columns = 4 #max number of columns displayed
```

```python
df
```

## Changing pandas Options with Methods

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 16)
```

```python
df
```

```python
pd.set_option("max_columns", 20)
```

```python
df
```

```python
pd.reset_option("max_rows")
```

```python
df
```

```python
pd.reset_option("max_columns")
```

```python
df
```

```python
pd.describe_option('max_columns')
```

```python
pd.describe_option('max_rows')
```

## The Precision Option-set # of digits for dataframe

```python
df = pd.DataFrame(np.random.randn(5,5))
df
```

```python
pd.get_option("precision")
```

```python
pd.set_option("precision", 2)
```

```python
df
```

```python
pd.reset_option("precision")
```

```python
df
```
