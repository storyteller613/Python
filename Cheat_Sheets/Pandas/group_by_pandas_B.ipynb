## The .agg() Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head()
```

```python
sectors.agg({"Revenue":["sum","mean"] ,
             "Profits":"sum",
             "Employees":"mean"})
```

```python
sectors.agg(["size","sum","mean"])
```

## Iterating through Groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
fortune.head()
```

```python
df = pd.DataFrame(columns = fortune.columns)
```

```python
for sector, data in sectors:
	highest_rev_co_in_group = data.nlargest(1, "Revenue")
	df = df.append(highest_rev_co_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
	highest_rev_in_city = data.nlargest(1, "Revenue")
	df = df.append(highest_rev_in_city)
```
