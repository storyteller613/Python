## The .unstack() Method

```python
world=pd.read_csv("worldstats.csv", index_col=["country","year"])
world.head(3)
```

```python
s = world.stack()
s.head()
```

```python
s.unstack().unstack().unstack()
```

```python
s.head(3)
```

```python
s.unstack(1).head(3) #moves the second row value (year) to the column values
```

```python
s.unstack(-1).head(3) #same as 2
```

```python
s.unstack("year").head(3)
```

```python
s.unstack(level=[1,0]) #outer layer year cause 1 first argument, second is country cause 0 second argument 
```

```python
s.unstack(level=["country","year"])
```

```python
s1 = s.unstack("year", fill_value=0)
s1.head()
```

## The pivot_table() Method

```python
food = pd.read_csv("foods.csv")
food.head(3)
```

```python
food.pivot_table(values='Spend', index='Gender', aggfunc='mean')
```

```python
food.pivot_table(values='Spend', index='Gender', aggfunc='sum')
```

```python
food.pivot_table(values='Spend', index='Item', aggfunc='sum')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], aggfunc='sum')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], columns='City',aggfunc='sum')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], columns=['Frequency','City'],aggfunc='sum')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], columns=['City'],aggfunc='count')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], columns=['City'],aggfunc='max')
```

```python
food.pivot_table(values='Spend', index=['Gender','Item'], columns=['City'],aggfunc='min').head()
```

```python
pd.pivot_table(data=food, values='Spend', index=['Gender','Item'], columns=['City'],aggfunc='min').head()
```
