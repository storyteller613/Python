## The .str.replace() Method

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail()
```

```python
"Hello World".replace("l","!")
```

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT","MANAGEMENT")
chicago.head()
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$","").astype("float")
chicago.head()
```

```python
chicago["Employee Annual Salary"].sum()
```

```python
chicago["Employee Annual Salary"].mean()
```

```python
chicago["Employee Annual Salary"].nlargest(10)
```

## Filtering with string methods: .contains(), .startswith(), .endswith()

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail()
```

```python
mask=chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
mask=chicago["Position Title"].str.lower().str.startswith("water")
chicago[mask]
```

```python
mask=chicago["Position Title"].str.lower().str.endswith("ist")
chicago[mask]
```

## More String Methods- .strip(), .lstrip(), and .rstrip(): remove white space from string

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail()
```

```python
"                Hello World    ".lstrip()
```

```python
"                Hello World    ".rstrip()
```

```python
"                Hello World    ".strip()
```

```python
chicago["Name"] = chicago["Name"].str.lstrip().str.rstrip()
```

```python
chicago["Position Title"] = chicago["Position Title"].str.strip()
```

## String methods on index and columns

```python
chicago = pd.read_csv("chicago.csv", index_col="Name").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail()
```

```python
chicago.index = chicago.index.str.strip().str.title()
chicago.head()
```

```python
chicago.columns = chicago.columns.str.upper()
chicago.head()
```

## Split Strings by Characters with .str.split() Method

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail()
```

```python
"Hello my name is Boris".split(" ")
```

```python
chicago["Name"].str.split(",").str.get(0).str.title().value_counts().head()
```

```python
chicago["Position Title"].str.split(" ").str.get(0).str.title().value_counts().head()
```

```python
chicago["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts().head()
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",", expand=True)
chicago.head()
```

```python
chicago[["First title word", "Remaining words"]] = chicago["Position Title"].str.split(" ", expand=True, n=1)
chicago.head()
```
