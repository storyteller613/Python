```python
bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_millions)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_millions)
```

## The .apply() Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def good_movie(row):
	actor = row[1]
	budget = row[4]
    
	if actor == "Pierce Brosnan":
		return "The Best"
	elif actor == "Roger Moore" and budget > 40:
		return "Enjoyable"
	else:
		return "I have no clue"
		
#bond.apply(good_movie, axis="columns")

bond["Movie Rank"] = bond.apply(func = good_movie, axis = 1)
bond.head()
```

```python
bond["Profit"] = bond["Box Office"] - bond["Budget"] - bond["Bond Actor Salary"]
bond["Profit Rank"] = "Null"

def profit_rank(row):
    if row["Profit"] <= 200:
        return "Lower"
    elif row["Profit"] > 200 and row["Profit"] <= 400:
        return "Medium"
    elif row["Profit"] > 400:
        return "High"
    else:
        return "NaN"
bond.apply(func = profit_rank, axis = 1)
bond["Profit Rank"] = bond.apply(func = profit_rank, axis = 1)
bond.head()
```
