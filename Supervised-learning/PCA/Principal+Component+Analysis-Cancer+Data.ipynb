# Principal Component Analysis With Cancer Data

## Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

## The Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#(['DESCR', 'data', 'feature_names', 'target_names', 'target'])
```

```python
df.head()
```

## PCA Visualization

```python
from sklearn.preprocessing import StandardScaler
```

```python
#scale data
scaler = StandardScaler()
scaler.fit(df)
```

```python
#scale data
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

## PCA: n_componenets=2

```python
pca_2 = PCA(n_components=2)
```

```python
pca_2.fit(scaled_data)
```

```python
x_pca_2 = pca_2.transform(scaled_data)
```

```python
x_pca_2.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca_2[:,0],x_pca_2[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Interpreting the components 

```python
pca_2.components_
```

```python
df_comp_2 = pd.DataFrame(pca_2.components_,columns=cancer['feature_names'])
```

```python
df_comp_2.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp_2,cmap='plasma')
```

### Analysis

This heatmap and the color bar basically represent the correlation between the various feature and the principal component itself.  The first principal component is more correlated with mean fractal dimension, fractal dimension error, and worst fractal dimension.  The second principal component is more correlated with mean concavity and mean concave points.
