## Compare annual stock price trends

```python
#import data
yahoo = pd.read_csv('C:/Users/Y/Documents/Python/Machine_learning/Time_Series/Time_Series/yahoo.csv')
yahoo.head()

# Convert the date column to datetime64
yahoo.date = pd.to_datetime(yahoo.date)

# Set date column as index, Use the inplace keyword to avoid copying the DataFrame.
yahoo.set_index('date', inplace=True)

# Inspect data 
print(data.info())

# Create dataframe prices here
prices = pd.DataFrame()

# Select data for each year and concatenate with prices here 
for year in ['2013', '2014', '2015']:
    price_per_year = yahoo.loc[year, ['price']].reset_index(drop=True)
    price_per_year.rename(columns={'price': year}, inplace=True)
    prices = pd.concat([prices, price_per_year], axis=1)

# Plot prices
prices.plot(subplots=True)
plt.show()
```

## Set and change time series frequency

```python
#import data
co = pd.read_csv('C:/Users/Y/Documents/Python/Machine_learning/Time_Series/Time_Series/co_cities.csv')
co.head()

# Convert the date column to datetime64
co.date = pd.to_datetime(co.date)

# Set date column as index, Use the inplace keyword to avoid copying the DataFrame.
co.set_index('date', inplace=True)

# Inspect data
print(co.info())

# Set the frequency to calendar daily
co = co.asfreq('D')

# Plot the data
co.plot(subplots=True)
plt.show()

# Set frequency to monthly
co = co.asfreq('M')

# Plot the data
co.plot(subplots=True)
plt.show()
```
